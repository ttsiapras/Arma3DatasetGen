```python
arucoDict = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_6X6_250)

for i in range(50):
    marker = cv2.aruco.generateImageMarker(arucoDict,i,512)
    
    cv2.imwrite(path%i,marker)
    markerB = cv2.copyMakeBorder(marker,
    top=50,
    bottom=50,
    left=50,
    right=50,
    borderType=cv2.BORDER_CONSTANT,
    value = 255)
    cv2.imwrite(pathB%i,markerB)

    cv2.imshow("marker",marker)
    cv2.imshow("markerB",markerB)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Find locations

```python
Cx,Cy,Cz = (1940.3,5568,3.2)
```

```python
#300m distance Upper targets z = 105 ####################################################################
pathB      = "images\\aruco%02dB.jpg"
F_item     = 2  # Initial Item
F_aruco    = 0   # Initial offset
F_id       = 2 # Initial offset

z           = 105
dist        = 300
for i,dir in enumerate(range(0,360,30)):    
    Ty = Cy+dist*m.cos(m.radians(dir+15)); Tx = Cx+dist*m.sin(m.radians(dir+15))

    item  = F_item+i
    aruco = pathB%(F_aruco+i)
    id    = F_id+i
    x = Tx
    y = Ty
    zangle = m.radians(dir+15)
    name = "aruco%02d"%(F_aruco+i)
    
    print(lines10%(item,x,z,y,zangle,name,id,aruco))

#300m distance Upper targets z = 35 ####################################################################
pathB      = "images\\aruco%02dB.jpg"
F_item     = F_item  + 12  # Initial Item
F_aruco    = F_aruco + 12   # Initial offset
F_id       = F_id + 12 # Initial offset

z           = 35
dist        = 300
for i,dir in enumerate(range(0,360,30)):    
    Ty = Cy+dist*m.cos(m.radians(dir+15)); Tx = Cx+dist*m.sin(m.radians(dir+15))

    item  = F_item+i
    aruco = pathB%(F_aruco+i)
    id    = F_id+i
    x = Tx
    y = Ty
    zangle = m.radians(dir+15)
    name = "aruco%02d"%(F_aruco+i)
    
    print(lines10%(item,x,z,y,zangle,name,id,aruco))

#30m distance Upper targets z = 35 ####################################################################
F_item     = F_item  + 12  # Initial Item
F_aruco    = F_aruco + 12   # Initial offset
F_id       = F_id + 12 # Initial offset

z           = 5
dist        = 30
for i,dir in enumerate(range(0,360,30)):    
    Ty = Cy+dist*m.cos(m.radians(dir+15)); Tx = Cx+dist*m.sin(m.radians(dir+15))

    item  = F_item+i
    aruco = pathB%(F_aruco+i)
    id    = F_id+i
    x = Tx
    y = Ty
    zangle = m.radians(dir+15)
    name = "aruco%02d"%(F_aruco+i)
    
    print(lines1%(item,x,z,y,zangle,name,id,aruco))
```
