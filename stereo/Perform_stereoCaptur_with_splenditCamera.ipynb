```python
import pyautogui
import pyperclip
import time
import ctypes
from ctypes import wintypes

#######################################################################
# Code borowed from Stackoverflow  https://stackoverflow.com/a/13615802
#######################################################################
user32 = ctypes.WinDLL('user32', use_last_error=True)

INPUT_MOUSE    = 0
INPUT_KEYBOARD = 1
INPUT_HARDWARE = 2

KEYEVENTF_EXTENDEDKEY = 0x0001
KEYEVENTF_KEYUP       = 0x0002
KEYEVENTF_UNICODE     = 0x0004
KEYEVENTF_SCANCODE    = 0x0008

MAPVK_VK_TO_VSC = 0

# C struct definitions

wintypes.ULONG_PTR = wintypes.WPARAM

class MOUSEINPUT(ctypes.Structure):
    _fields_ = (("dx",          wintypes.LONG),
                ("dy",          wintypes.LONG),
                ("mouseData",   wintypes.DWORD),
                ("dwFlags",     wintypes.DWORD),
                ("time",        wintypes.DWORD),
                ("dwExtraInfo", wintypes.ULONG_PTR))

class KEYBDINPUT(ctypes.Structure):
    _fields_ = (("wVk",         wintypes.WORD),
                ("wScan",       wintypes.WORD),
                ("dwFlags",     wintypes.DWORD),
                ("time",        wintypes.DWORD),
                ("dwExtraInfo", wintypes.ULONG_PTR))

    def __init__(self, *args, **kwds):
        super(KEYBDINPUT, self).__init__(*args, **kwds)
        # some programs use the scan code even if KEYEVENTF_SCANCODE
        # isn't set in dwFflags, so attempt to map the correct code.
        if not self.dwFlags & KEYEVENTF_UNICODE:
            self.wScan = user32.MapVirtualKeyExW(self.wVk,
                                                 MAPVK_VK_TO_VSC, 0)

class HARDWAREINPUT(ctypes.Structure):
    _fields_ = (("uMsg",    wintypes.DWORD),
                ("wParamL", wintypes.WORD),
                ("wParamH", wintypes.WORD))

class INPUT(ctypes.Structure):
    class _INPUT(ctypes.Union):
        _fields_ = (("ki", KEYBDINPUT),
                    ("mi", MOUSEINPUT),
                    ("hi", HARDWAREINPUT))
    _anonymous_ = ("_input",)
    _fields_ = (("type",   wintypes.DWORD),
                ("_input", _INPUT))

LPINPUT = ctypes.POINTER(INPUT)

def _check_count(result, func, args):
    if result == 0:
        raise ctypes.WinError(ctypes.get_last_error())
    return args

user32.SendInput.errcheck = _check_count
user32.SendInput.argtypes = (wintypes.UINT, # nInputs
                             LPINPUT,       # pInputs
                             ctypes.c_int)  # cbSize

class VK_code():
    def __init__(self,name,hexCode,description):
        self.value = hexCode
        self.description = description
        self.name  = name

    def __str__(self):
        return(self.name + ": " + hex(self.value)+ ": " +self.description)

# https://learn.microsoft.com/en-us/windows/win32/inputdev/virtual-key-codes

VK_ABNT_C1	     = VK_code("VK_ABNT_C1	     "          , 0xC1, "Abnt C1")
VK_ABNT_C2	     = VK_code("VK_ABNT_C2	     "          , 0xC2, "Abnt C2")
VK_ADD	         = VK_code("VK_ADD	         "          , 0x6B, "Numpad +")
VK_ATTN	         = VK_code("VK_ATTN	         "          , 0xF6, "Attn")
VK_BACK	         = VK_code("VK_BACK	         "          , 0x08, "Backspace")
VK_CANCEL	     = VK_code("VK_CANCEL	     "          , 0x03, "Break")
VK_CLEAR	     = VK_code("VK_CLEAR	     "          , 0x0C, "Clear")
VK_CRSEL	     = VK_code("VK_CRSEL	     "          , 0xF7, "Cr Sel")
VK_DECIMAL	     = VK_code("VK_DECIMAL	     "          , 0x6E, "Numpad .")
VK_DIVIDE	     = VK_code("VK_DIVIDE	     "          , 0x6F, "Numpad /")
VK_EREOF	     = VK_code("VK_EREOF	     "          , 0xF9, "Er Eof")
VK_ESCAPE	     = VK_code("VK_ESCAPE	     "          , 0x1B, "Esc")
VK_EXECUTE	     = VK_code("VK_EXECUTE	     "          , 0x2B, "Execute")
VK_EXSEL	     = VK_code("VK_EXSEL	     "          , 0xF8, "Ex Sel")
VK_ICO_CLEAR     = VK_code("VK_ICO_CLEAR     "          , 0xE6, "IcoClr")
VK_ICO_HELP	     = VK_code("VK_ICO_HELP	     "          , 0xE3, "IcoHlp")
VK_KEY_0	     = VK_code("VK_KEY_0	     "          , 0x30, "0")
VK_KEY_1	     = VK_code("VK_KEY_1	     "          , 0x31, "1")
VK_KEY_2	     = VK_code("VK_KEY_2	     "          , 0x32, "2")
VK_KEY_3	     = VK_code("VK_KEY_3	     "          , 0x33, "3")
VK_KEY_4	     = VK_code("VK_KEY_4	     "          , 0x34, "4")
VK_KEY_5	     = VK_code("VK_KEY_5	     "          , 0x35, "5")
VK_KEY_6	     = VK_code("VK_KEY_6	     "          , 0x36, "6")
VK_KEY_7	     = VK_code("VK_KEY_7	     "          , 0x37, "7")
VK_KEY_8	     = VK_code("VK_KEY_8	     "          , 0x38, "8")
VK_KEY_9	     = VK_code("VK_KEY_9	     "          , 0x39, "9")
VK_KEY_A	     = VK_code("VK_KEY_A	     "          , 0x41, "A")
VK_KEY_B	     = VK_code("VK_KEY_B	     "          , 0x42, "B")
VK_KEY_C	     = VK_code("VK_KEY_C	     "          , 0x43, "C")
VK_KEY_D	     = VK_code("VK_KEY_D	     "          , 0x44, "D")
VK_KEY_E	     = VK_code("VK_KEY_E	     "          , 0x45, "E")
VK_KEY_F	     = VK_code("VK_KEY_F	     "          , 0x46, "F")
VK_KEY_G	     = VK_code("VK_KEY_G	     "          , 0x47, "G")
VK_KEY_H	     = VK_code("VK_KEY_H	     "          , 0x48, "H")
VK_KEY_I	     = VK_code("VK_KEY_I	     "          , 0x49, "I")
VK_KEY_J	     = VK_code("VK_KEY_J	     "          , 0x4A, "J")
VK_KEY_K	     = VK_code("VK_KEY_K	     "          , 0x4B, "K")
VK_KEY_L	     = VK_code("VK_KEY_L	     "          , 0x4C, "L")
VK_KEY_M	     = VK_code("VK_KEY_M	     "          , 0x4D, "M")
VK_KEY_N	     = VK_code("VK_KEY_N	     "          , 0x4E, "N")
VK_KEY_O	     = VK_code("VK_KEY_O	     "          , 0x4F, "O")
VK_KEY_P	     = VK_code("VK_KEY_P	     "          , 0x50, "P")
VK_KEY_Q	     = VK_code("VK_KEY_Q	     "          , 0x51, "Q")
VK_KEY_R	     = VK_code("VK_KEY_R	     "          , 0x52, "R")
VK_KEY_S	     = VK_code("VK_KEY_S	     "          , 0x53, "S")
VK_KEY_T	     = VK_code("VK_KEY_T	     "          , 0x54, "T")
VK_KEY_U	     = VK_code("VK_KEY_U	     "          , 0x55, "U")
VK_KEY_V	     = VK_code("VK_KEY_V	     "          , 0x56, "V")
VK_KEY_W	     = VK_code("VK_KEY_W	     "          , 0x57, "W")
VK_KEY_X	     = VK_code("VK_KEY_X	     "          , 0x58, "X")
VK_KEY_Y	     = VK_code("VK_KEY_Y	     "          , 0x59, "Y")
VK_KEY_Z	     = VK_code("VK_KEY_Z	     "          , 0x5A, "Z")
VK_MULTIPLY	     = VK_code("VK_MULTIPLY	     "          , 0x6A, "Numpad *")
VK_NONAME	     = VK_code("VK_NONAME	     "          , 0xFC, "NoName")
VK_NUMPAD0	     = VK_code("VK_NUMPAD0	     "          , 0x60, "Numpad 0")
VK_NUMPAD1	     = VK_code("VK_NUMPAD1	     "          , 0x61, "Numpad 1")
VK_NUMPAD2	     = VK_code("VK_NUMPAD2	     "          , 0x62, "Numpad 2")
VK_NUMPAD3	     = VK_code("VK_NUMPAD3	     "          , 0x63, "Numpad 3")
VK_NUMPAD4	     = VK_code("VK_NUMPAD4	     "          , 0x64, "Numpad 4")
VK_NUMPAD5	     = VK_code("VK_NUMPAD5	     "          , 0x65, "Numpad 5")
VK_NUMPAD6	     = VK_code("VK_NUMPAD6	     "          , 0x66, "Numpad 6")
VK_NUMPAD7	     = VK_code("VK_NUMPAD7	     "          , 0x67, "Numpad 7")
VK_NUMPAD8	     = VK_code("VK_NUMPAD8	     "          , 0x68, "Numpad 8")
VK_NUMPAD9	     = VK_code("VK_NUMPAD9	     "          , 0x69, "Numpad 9")
VK_OEM_1	     = VK_code("VK_OEM_1	     "          , 0xBA, "OEM_1 (: ;)")
VK_OEM_102	     = VK_code("VK_OEM_102	     "          , 0xE2, "OEM_102 (> <)")
VK_OEM_2	     = VK_code("VK_OEM_2	     "          , 0xBF, "OEM_2 (? /)")
VK_OEM_3	     = VK_code("VK_OEM_3	     "          , 0xC0, "OEM_3 (~ `)")
VK_OEM_4	     = VK_code("VK_OEM_4	     "          , 0xDB, "OEM_4 ({ [)")
VK_OEM_5	     = VK_code("VK_OEM_5	     "          , 0xDC, "OEM_5 (| \\)")
VK_OEM_6	     = VK_code("VK_OEM_6	     "          , 0xDD, "OEM_6 (} ])")
VK_OEM_7	     = VK_code("VK_OEM_7	     "          , 0xDE, "OEM_7 (\" \')")
VK_OEM_8	     = VK_code("VK_OEM_8	     "          , 0xDF, "OEM_8 (ยง !)")
VK_OEM_ATTN	     = VK_code("VK_OEM_ATTN	     "          , 0xF0, "Oem Attn")
VK_OEM_AUTO	     = VK_code("VK_OEM_AUTO	     "          , 0xF3, "Auto")
VK_OEM_AX	     = VK_code("VK_OEM_AX	     "          , 0xE1, "Ax")
VK_OEM_BACKTAB	 = VK_code("VK_OEM_BACKTAB	 "          , 0xF5, "Back Tab")
VK_OEM_CLEAR	 = VK_code("VK_OEM_CLEAR	 "          , 0xFE, "OemClr")
VK_OEM_COMMA	 = VK_code("VK_OEM_COMMA	 "          , 0xBC, "OEM_COMMA (< ,)")
VK_OEM_COPY	     = VK_code("VK_OEM_COPY	     "          , 0xF2, "Copy")
VK_OEM_CUSEL	 = VK_code("VK_OEM_CUSEL	 "          , 0xEF, "Cu Sel")
VK_OEM_ENLW	     = VK_code("VK_OEM_ENLW	     "          , 0xF4, "Enlw")
VK_OEM_FINISH	 = VK_code("VK_OEM_FINISH	 "          , 0xF1, "Finish")
VK_OEM_FJ_LOYA	 = VK_code("VK_OEM_FJ_LOYA	 "          , 0x95, "Loya")
VK_OEM_FJ_MASSHOU= VK_code("VK_OEM_FJ_MASSHOU"          , 0x93, "Mashu")
VK_OEM_FJ_ROYA	 = VK_code("VK_OEM_FJ_ROYA	 "          , 0x96, "Roya")
VK_OEM_FJ_TOUROKU= VK_code("VK_OEM_FJ_TOUROKU"          , 0x94, "Touroku")
VK_OEM_JUMP	     = VK_code("VK_OEM_JUMP	     "          , 0xEA, "Jump")
VK_OEM_MINUS	 = VK_code("VK_OEM_MINUS	 "          , 0xBD, "OEM_MINUS (_ -)")
VK_OEM_PA1	     = VK_code("VK_OEM_PA1	     "          , 0xEB, "OemPa1")
VK_OEM_PA2	     = VK_code("VK_OEM_PA2	     "          , 0xEC, "OemPa2")
VK_OEM_PA3	     = VK_code("VK_OEM_PA3	     "          , 0xED, "OemPa3")
VK_OEM_PERIOD	 = VK_code("VK_OEM_PERIOD	 "          , 0xBE, "OEM_PERIOD (> .)")
VK_OEM_PLUS	     = VK_code("VK_OEM_PLUS	     "          , 0xBB, "OEM_PLUS (+ =)")
VK_OEM_RESET	 = VK_code("VK_OEM_RESET	 "          , 0xE9, "Reset")
VK_OEM_WSCTRL	 = VK_code("VK_OEM_WSCTRL	 "          , 0xEE, "WsCtrl")
VK_PA1	         = VK_code("VK_PA1	         "          , 0xFD, "Pa1")
VK_PACKET	     = VK_code("VK_PACKET	     "          , 0xE7, "Packet")
VK_PLAY	         = VK_code("VK_PLAY	         "          , 0xFA, "Play")
VK_PROCESSKEY	 = VK_code("VK_PROCESSKEY	 "          , 0xE5, "Process")
VK_RETURN	     = VK_code("VK_RETURN	     "          , 0x0D, "Enter")
VK_SELECT	     = VK_code("VK_SELECT	     "          , 0x29, "Select")
VK_SEPARATOR	 = VK_code("VK_SEPARATOR	 "          , 0x6C, "Separator")
VK_SPACE	     = VK_code("VK_SPACE	     "          , 0x20, "Space")
VK_SUBTRACT	     = VK_code("VK_SUBTRACT	     "          , 0x6D, "Num -")
VK_TAB	         = VK_code("VK_TAB	         "          , 0x09, "Tab")
VK_ZOOM	         = VK_code("VK_ZOOM	         "          , 0xFB, "Zoom")
VK__none_	          = VK_code("VK__none_	           ", 0xFF, "no VK mapping")
VK_ACCEPT	          = VK_code("VK_ACCEPT	           ", 0x1E, "Accept")
VK_APPS	              = VK_code("VK_APPS	           ", 0x5D, "Context Menu")
VK_BROWSER_BACK	      = VK_code("VK_BROWSER_BACK	   ", 0xA6, "Browser Back")
VK_BROWSER_FAVORITES  = VK_code("VK_BROWSER_FAVORITES  ", 0xAB, "Browser Favorites")
VK_BROWSER_FORWARD	  = VK_code("VK_BROWSER_FORWARD	   ", 0xA7, "Browser Forward")
VK_BROWSER_HOME	      = VK_code("VK_BROWSER_HOME	   ", 0xAC, "Browser Home")
VK_BROWSER_REFRESH	  = VK_code("VK_BROWSER_REFRESH	   ", 0xA8, "Browser Refresh")
VK_BROWSER_SEARCH	  = VK_code("VK_BROWSER_SEARCH	   ", 0xAA, "Browser Search")
VK_BROWSER_STOP 	  = VK_code("VK_BROWSER_STOP 	   ", 0xA9, "Browser Stop")
VK_CAPITAL	          = VK_code("VK_CAPITAL	           ", 0x14, "Caps Lock")
VK_CONVERT	          = VK_code("VK_CONVERT	           ", 0x1C, "Convert")
VK_DELETE	          = VK_code("VK_DELETE	           ", 0x2E, "Delete")
VK_DOWN	              = VK_code("VK_DOWN	           ", 0x28, "Arrow Down")
VK_END	              = VK_code("VK_END	               ", 0x23, "End")
VK_F1	              = VK_code("VK_F1	               ", 0x70, "F1")
VK_F10	              = VK_code("VK_F10	               ", 0x79, "F10")
VK_F11	              = VK_code("VK_F11	               ", 0x7A, "F11")
VK_F12	              = VK_code("VK_F12	               ", 0x7B, "F12")
VK_F13	              = VK_code("VK_F13	               ", 0x7C, "F13")
VK_F14	              = VK_code("VK_F14	               ", 0x7D, "F14")
VK_F15	              = VK_code("VK_F15	               ", 0x7E, "F15")
VK_F16	              = VK_code("VK_F16	               ", 0x7F, "F16")
VK_F17	              = VK_code("VK_F17	               ", 0x80, "F17")
VK_F18	              = VK_code("VK_F18	               ", 0x81, "F18")
VK_F19	              = VK_code("VK_F19	               ", 0x82, "F19")
VK_F2	              = VK_code("VK_F2	               ", 0x71, "F2")
VK_F20	              = VK_code("VK_F20	               ", 0x83, "F20")
VK_F21	              = VK_code("VK_F21	               ", 0x84, "F21")
VK_F22	              = VK_code("VK_F22	               ", 0x85, "F22")
VK_F23	              = VK_code("VK_F23	               ", 0x86, "F23")
VK_F24	              = VK_code("VK_F24	               ", 0x87, "F24")
VK_F3	              = VK_code("VK_F3	               ", 0x72, "F3")
VK_F4	              = VK_code("VK_F4	               ", 0x73, "F4")
VK_F5	              = VK_code("VK_F5	               ", 0x74, "F5")
VK_F6	              = VK_code("VK_F6	               ", 0x75, "F6")
VK_F7	              = VK_code("VK_F7	               ", 0x76, "F7")
VK_F8	              = VK_code("VK_F8	               ", 0x77, "F8")
VK_F9	              = VK_code("VK_F9	               ", 0x78, "F9")
VK_FINAL	          = VK_code("VK_FINAL	           ", 0x18, "Final")
VK_HELP	              = VK_code("VK_HELP	           ", 0x2F, "Help")
VK_HOME	              = VK_code("VK_HOME	           ", 0x24, "Home")
VK_ICO_00	          = VK_code("VK_ICO_00	           ", 0xE4, "Ico00 *")
VK_INSERT	          = VK_code("VK_INSERT	           ", 0x2D, "Insert")
VK_JUNJA	          = VK_code("VK_JUNJA	           ", 0x17, "Junja")
VK_KANA	              = VK_code("VK_KANA	           ", 0x15, "Kana")
VK_KANJI	          = VK_code("VK_KANJI	           ", 0x19, "Kanji")
VK_LAUNCH_APP1	      = VK_code("VK_LAUNCH_APP1	       ", 0xB6, "App1")
VK_LAUNCH_APP2	      = VK_code("VK_LAUNCH_APP2	       ", 0xB7, "App2")
VK_LAUNCH_MAIL	      = VK_code("VK_LAUNCH_MAIL	       ", 0xB4, "Mail")
VK_LAUNCH_MEDIA_SELECT= VK_code("VK_LAUNCH_MEDIA_SELECT", 0xB5, "Media")
VK_LBUTTON	          = VK_code("VK_LBUTTON	           ", 0x01, "Left Button **")
VK_LCONTROL	          = VK_code("VK_LCONTROL	       ", 0xA2, "Left Ctrl")
VK_LEFT	              = VK_code("VK_LEFT	           ", 0x25, "Arrow Left")
VK_LMENU	          = VK_code("VK_LMENU	           ", 0xA4, "Left Alt")
VK_LSHIFT	          = VK_code("VK_LSHIFT	           ", 0xA0, "Left Shift")
VK_LWIN	              = VK_code("VK_LWIN	           ", 0x5B, "Left Win")
VK_MBUTTON	          = VK_code("VK_MBUTTON	           ", 0x04, "Middle Button **")
VK_MEDIA_NEXT_TRACK   = VK_code("VK_MEDIA_NEXT_TRACK   ", 0xB0, "Next Track")
VK_MEDIA_PLAY_PAUSE   = VK_code("VK_MEDIA_PLAY_PAUSE   ", 0xB3, "Play / Pause")
VK_MEDIA_PREV_TRACK   = VK_code("VK_MEDIA_PREV_TRACK   ", 0xB1, "Previous Track")
VK_MEDIA_STOP	      = VK_code("VK_MEDIA_STOP	       ", 0xB2, "Stop")
VK_MODECHANGE	      = VK_code("VK_MODECHANGE	       ", 0x1F, "Mode Change")
VK_NEXT	              = VK_code("VK_NEXT	           ", 0x22, "Page Down")
VK_NONCONVERT	      = VK_code("VK_NONCONVERT	       ", 0x1D, "Non Convert")
VK_NUMLOCK	          = VK_code("VK_NUMLOCK	           ", 0x90, "Num Lock")
VK_OEM_FJ_JISHO	      = VK_code("VK_OEM_FJ_JISHO	   ", 0x92, "Jisho")
VK_PAUSE	          = VK_code("VK_PAUSE	           ", 0x13, "Pause")
VK_PRINT	          = VK_code("VK_PRINT	           ", 0x2A, "Print")
VK_PRIOR	          = VK_code("VK_PRIOR	           ", 0x21, "Page Up")
VK_RBUTTON	          = VK_code("VK_RBUTTON	           ", 0x02, "Right Button **")
VK_RCONTROL	          = VK_code("VK_RCONTROL	       ", 0xA3, "Right Ctrl")
VK_RIGHT	          = VK_code("VK_RIGHT	           ", 0x27, "Arrow Right")
VK_RMENU	          = VK_code("VK_RMENU	           ", 0xA5, "Right Alt")
VK_RSHIFT	          = VK_code("VK_RSHIFT	           ", 0xA1, "Right Shift")
VK_RWIN	              = VK_code("VK_RWIN	           ", 0x5C, "Right Win")
VK_SCROLL	          = VK_code("VK_SCROLL	           ", 0x91, "Scrol Lock")
VK_SLEEP	          = VK_code("VK_SLEEP	           ", 0x5F, "Sleep")
VK_SNAPSHOT	          = VK_code("VK_SNAPSHOT	       ", 0x2C, "Print Screen")
VK_UP	              = VK_code("VK_UP	               ", 0x26, "Arrow Up")
VK_VOLUME_DOWN	      = VK_code("VK_VOLUME_DOWN	       ", 0xAE, "Volume Down")
VK_VOLUME_MUTE	      = VK_code("VK_VOLUME_MUTE	       ", 0xAD, "Volume Mute")
VK_VOLUME_UP	      = VK_code("VK_VOLUME_UP	       ", 0xAF, "Volume Up")
VK_XBUTTON1	          = VK_code("VK_XBUTTON1	       ", 0x05, "X Button 1 **")
VK_XBUTTON2	          = VK_code("VK_XBUTTON2	       ", 0x06, "X Button 2 **")

# Functions

def PressKey(hexKeyCode):
    x = INPUT(type=INPUT_KEYBOARD,
              ki=KEYBDINPUT(wVk=hexKeyCode))
    user32.SendInput(1, ctypes.byref(x), ctypes.sizeof(x))

def ReleaseKey(hexKeyCode):
    x = INPUT(type=INPUT_KEYBOARD,
              ki=KEYBDINPUT(wVk=hexKeyCode,
                            dwFlags=KEYEVENTF_KEYUP))
    user32.SendInput(1, ctypes.byref(x), ctypes.sizeof(x))\
    
# Cntrl-V and other key  press funcions from pyautogui could not work on ARMA3
# while working normally in other apps link notepad for example.
# This is why i use this method on keyboard input for ARMA3
def ARMA_CntrV():
    PressKey(VK_LCONTROL.value) 
    PressKey(VK_KEY_V.value)   
    ReleaseKey(VK_KEY_V.value) 
    time.sleep(0.1)
    ReleaseKey(VK_LCONTROL.value)

def ARMA_ChangeSensor():
    PressKey(VK_KEY_N.value)   
    ReleaseKey(VK_KEY_N.value) 

def ARMA_toggle_HUD():
    PressKey(VK_BACK.value) 
    time.sleep(0.05)
    ReleaseKey(VK_BACK.value)

def ARMA_steam_screenshot():
    PressKey(VK_F12.value) 
    time.sleep(0.05)
    ReleaseKey(VK_F12.value)


# This class include information about the array describing the <splendit camera> tool and 
# can be copied using Cntrl-C when in the <splendit camera> window.
class camera_string():
    def __init__(self,initialStr):
        self.map = initialStr[0]
        self.position = initialStr[1]
        self.cam_Yaw  = initialStr[2]
        self.FoV      = initialStr[3]
        self.camPR    = initialStr[4]
        self.focus    = initialStr[5]
        self.aperture = initialStr[6]
        self.daytime  = initialStr[7]
        self.overcast = initialStr[8]
        self.acctime  = initialStr[9]
        self.bright   = initialStr[10]
        self.contrast = initialStr[11]
        self.saturate = initialStr[12]

    def __str__(self):
        outputStr = '['
        outputStr += "\"" + self.map +  "\"" + ','
        outputStr += str(self.position  ) + ','
        outputStr += str(self.cam_Yaw   ) + ','
        outputStr += str(self.FoV       ) + ','
        outputStr += str(self.camPR     ) + ','
        outputStr += str(self.focus     ) + ','
        outputStr += str(self.aperture  ) + ','
        outputStr += str(self.daytime   ) + ','
        outputStr += str(self.overcast  ) + ','
        outputStr += str(self.acctime   ) + ','
        outputStr += str(self.bright    ) + ','
        outputStr += str(self.contrast  ) + ','                                                       
        outputStr += str(self.saturate  ) + ']'

        return outputStr
```

```python
import cv2
import os

# Folder with images
image_folder = r"C:\Users\user1\Documents\Repositories\Arma3DatasetGen\stereo\images\odd"
# Output video name
video_name = 'output_raw_odd.avi'

# Get and sort images
images = [img for img in os.listdir(image_folder) if img.endswith(('.png', '.jpg', '.jpeg'))]
images.sort()

# Read the first frame to get dimensions
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

# Create VideoWriter object with RAW codec
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # <- RAW video
video = cv2.VideoWriter(video_name, fourcc, 10, (width, height))  # 30 FPS

for image in images:
    frame = cv2.imread(os.path.join(image_folder, image))
    video.write(frame)

video.release()
print(f"RAW video saved as {video_name}")

```

```python
import os
import shutil

# --- CONFIGURATION ---
input_folder = r'D:\arma3_screenshots'
output_folder_odd = 'images/odd'
output_folder_even = 'images/even'

start_file = '107410_20250425170757_1.png'
end_file = '107410_20250425173121_1.png'

# --- MAKE SURE OUTPUT FOLDERS EXIST ---
os.makedirs(output_folder_odd, exist_ok=True)
os.makedirs(output_folder_even, exist_ok=True)

# --- LOAD AND SORT FILES ---
all_files = sorted([f for f in os.listdir(input_folder) if f.endswith('.png')])

# --- FIND RANGE ---
try:
    start_idx = all_files.index(start_file)
    end_idx = all_files.index(end_file)
except ValueError:
    print("Start or end file not found!")
    exit(1)

# Ensure correct order
if start_idx > end_idx:
    start_idx, end_idx = end_idx, start_idx

selected_files = all_files[start_idx:end_idx + 1]

# --- PROCESS FILES ---
for idx, filename in enumerate(selected_files):
    src = os.path.join(input_folder, filename)
    dst_folder = output_folder_odd if idx % 2 else output_folder_even
    shutil.copy(src, dst_folder)

print("Processing done!")
print(end_idx-start_idx)
```
